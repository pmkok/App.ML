```python
from skimage.io import imshow, imread
from skimage.feature import hog
from skimage import exposure
import skimage.transform as transform
import skimage.io
import skimage.exposure
import cv2
```

```python
img = imread('van.jpeg', as_gray=True)

# Extrahiere HOG-Merkmale
features, hog_image = hog(img, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2), visualize=True)

# Skaliere das HOG-Bild
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 5))

# Zeige das skalierte HOG-Bild an
imshow(hog_image_rescaled)

# Speichere die HOG-Merkmale in einem Array
hog_features_array = features.reshape(-1) # Das Array wird zu einer eindimensionalen Form umgewandelt

print('HOG-Merkmale:', hog_features_array)
```

```python
# Bild einlesen
img = skimage.io.imread('haaland.jpg', plugin='matplotlib')

# Graustufen-Konvertierung
img_gray = skimage.color.rgb2gray(img)

# Bildgröße anpassen
reshaped_img = transform.resize(img_gray, (240, 240))

# Normalisierung des Bildes
normalized_image = skimage.exposure.rescale_intensity(reshaped_img, in_range='image', out_range=(0, 1))

# HOG-Merkmale extrahieren
hog_features, hog_image = hog(normalized_image, orientations=8, pixels_per_cell=(4, 4), cells_per_block=(2, 2), visualize=True)

# Skalierung des HOG-Bildes
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 5))

# HOG-Bild anzeigen
skimage.io.imshow(hog_image_rescaled)

# HOG-Merkmale in ein Array speichern
hog_features_array = hog_features.reshape(-1)

print('HOG-Merkmale:', hog_features_array)
```
